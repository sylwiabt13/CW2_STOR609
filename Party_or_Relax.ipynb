# Party or Relax, a simple example MDP

Suppose you are a uni student deciding on whether you should party or rest on every weekend, you enjoy partying much more than resting, but you are seriously worried about the risk of getting sick. The example we go through is portrayed in example 9.27 here https://artint.info/2e/html2e/ArtInt2e.Ch9.S5.html#Ch9.Thmciexamplered27 .

You are either healthy or sick, therefore the corresponding state space is $\mathcal S= \{\text{Healthy},\text{Sick}\}$.

On each weekend you decide to either relax or party, therefore the corresponding action space is $\mathcal A = \{\text{Relax},\text{Party}\}$

You then elicit the probabilities $P(s'\mid s,a)$ you are healthy or sick given last weekend you were and did $(s,a)\in \mathcal S \times \mathcal A$:

| State   | Action | Next State | Probability |
| ------- | ------ | ---------- | ----------- |
| healthy | relax  | healthy    | 0.95        |
| healthy | relax  | sick       | 0.05        |
| healthy | party  | healthy    | 0.70        |
| healthy | party  | sick       | 0.30        |
| sick    | relax  | healthy    | 0.50        |
| sick    | relax  | sick       | 0.50        |
| sick    | party  | healthy    | 0.10        |
| sick    | party  | sick       | 0.90        |

You prefer to party over relaxing, but you do not like being sick at all, you characterise your preferences of each state-action pair $R(s,a)$ in the following table:

| State   | Action | Reward |
| ------- | ------ | ------ |
| healthy | relax  | 7      |
| healthy | party  | 10     |
| sick    | relax  | 0      |
| sick    | party  | 2      |

To solve this problem and find the optimal policy, we use value iteration:

```python
S = ['Healthy','Sick'] #Healthy or Sick
A = ['Relax','Party'] #Relax or Party
P={('Healthy','Relax'):[0.95,0.05],('Healthy','Party'):[0.7,0.3],('Sick','Relax'):[0.5,0.5],('Sick','Party'):[0.1,0.9]} #State transition probabilities
R={('Healthy','Relax'):7,('Healthy','Party'):10,('Sick','Relax'):0,('Sick','Party'):2} #Reward probabilities

from VIpackage import value_iteration
```

The state and action spaces are inputed as lists of states, whereas we express the transition probabilities and rewards as dictionaries which characterise the functions.

To carry out value iteration you need to select a discount factor $\gamma\in [0,1]$, this is a measure of how much you care about the future rewards, having a $\gamma$ closer to zero means you care more about what is happening in the current iteration of the algorithm, whereas if $\gamma$ is closer to one suggests you care more about the future rewards, and the corresponding policy will seek to maximise the cumulative reward in the long-run.

```python
value_iteration(S, A, P, R, 0.99, 2000)
#ALWAYS RELAX
```

```python
value_iteration(S, A, P, R, 0.5, 1000)
#PARTY IF HEALTHY
```

```python
value_iteration(S, A, P, R, 0.1, 1000)
#ALWAYS PARTY
```

As we can see, for a large value of $\gamma$, we prioritise long-term rewards and the corresponding policy is to always RELAX. If we decrease $\gamma$, the corresponding optimal policy becomes PARTY only if we are HEALTHY, and eventually for a low enough value of $\gamma$, the optimal policy becomes to always PARTY. This reflects how $\gamma$ characterises short-term and long-term behaviour. Note as the algorithms have run for a different number of iterations, we cannot compare the output of the value functions in this setting.

We now will design a unit-test based on this problem, by manually calculating 2 iterations of the algorithm for a given $\gamma$. Starting with $V_0(h)=V_0(s)=0$, we have for the first iteration that:

$$\begin{align*}
Q_1(h,r)&=7 \\
Q_1(h,p)&=10\\
Q_1(s,r)&=0\\
Q_1(s,p)&=2\\
V_1(h)&=10\\
V_1(s)&=2\\
\end{align*}$$


The second iteration then becomes:


$$\begin{align*}
Q_2(h,r)&=7 +9.6\gamma \\
Q_2(h,p)&=10+7.6\gamma\\
Q_2(s,r)&=6\gamma\\
Q_2(s,p)&=2+2.8\gamma\\
V_2(h)&=10+7.6\gamma\\
V_2(s)&=\max\{2+2.8\gamma,6\gamma\}\\
\end{align*}$$

Our code agrees with our calculations:

```python
value_iteration(S, A, P, R, 1, 2)
```
